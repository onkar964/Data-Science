```python
df_date_time  = pd.DataFrame({ "DT" : ("2018-12-28 12:24:59" , 
                                       "2019-05-9  2:10:25" ,
                                       "2019-12-31  22:10:25" ,
                                       "2017-01-01  9:01:20",
                                       "2017-07-03  22:10:25") } )
df_date_time
```

```python
df_date_time.info()
```

```python
df_date_time.DT = pd.to_datetime(df_date_time.DT)
df_date_time['Year']  = df_date_time.DT.dt.year
df_date_time['Month']= df_date_time.DT.dt.month
df_date_time['Day']= df_date_time.DT.dt.day
df_date_time['Hours']= df_date_time.DT.dt.hour
df_date_time['minutes']= df_date_time.DT.dt.minute
df_date_time['seconds']= df_date_time.DT.dt.second
df_date_time['Day of week']= df_date_time.DT.dt.day_name()
df_date_time['day of year']= df_date_time.DT.dt.dayofyear
```
