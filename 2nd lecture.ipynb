```python
def addnum(a,b):
    """
    this is documentation string
    """
    try:
        print(a)
        print(b)
        global result
        result = a+b
        print("result ", result)
    except:
        print("you have made error bocz recheck enter data")
    
    finally:
        print("this is finally print both condition when error is present or not")
```
